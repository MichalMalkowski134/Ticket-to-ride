```python
from PIL import Image
from os import walk
w = walk("Nowa---heksy-2")
for (dirpath, dirnames, filenames) in w:
  if not dirnames:
    if dirpath[-6:] == 'images':
      for name in filenames:
        im = Image.open(dirpath + '/' + name)
        f = open(dirpath[:-6] + '/labels/' + name[:-4] + '.txt', "r")
        lines = f.readlines()
        counter = 0
        for i in lines:
          line = i.split(' ')
          classname = line[0]
          x = float(line[1])
          y = float(line[2])
          w = float(line[3])
          h = float(line[4])
          cropped_im = im.crop(((x-w/2)*im.size[0], (y-h/2)*im.size[1], (x+w/2)*im.size[0], (y+h/2)*im.size[1]))
          cropped_im.save(dirpath + '/' + 'cropped_' + str(counter) + name)
          counter+=1
          f2 = open(dirpath[:-6] + '/labels/' + 'cropped_' + str(counter) + name[:-4] + '.txt', "w")
          f2.write(classname + ' 0.5 0.5 0.99 0.99')




```
